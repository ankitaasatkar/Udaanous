## visalisation

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Bar Chart

```python
data = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5, 6],
    'churn_status': ['Churned', 'Not Churned', 'Churned', 'Churned', 'Not Churned', 'Churned'],
    'region': ['East', 'West', 'East', 'West', 'West', 'East']  # cat_col_2
})
```

```python
cross_tab = pd.crosstab(data['region'], data['churn_status'])
```

```python
plt.figure(figsize=(8, 5)) 
cross_tab.plot(kind='bar', stacked=False)
plt.title('Churn Status by Region')
plt.xlabel('Churn Status')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='churn_status', hue='region', data=data)
```

## Stacked or clustered

```python
data = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5, 6, 7, 8],
    'churn_status': ['Churned', 'Not Churned', 'Churned', 'Not Churned',
                     'Churned', 'Not Churned', 'Churned', 'Not Churned'],
    'region': ['East', 'East', 'West', 'West', 'East', 'West', 'East', 'West']
})
```

```python
cross_tab = pd.crosstab(data['churn_status'], data['region'])


plt.figure(figsize=(8, 5))  
cross_tab.plot(kind='bar', stacked=True) 
plt.title('Churn Status by Region (Stacked)')
plt.xlabel('Churn Status')
plt.ylabel('Number of Customers')
plt.show()
```

```python
sns.countplot(x='churn_status', hue='region', data=data)
```

## pie chart

```python
data = pd.DataFrame({
    'customer_id': [1, 2, 3, 4, 5, 6],
    'churn_status': ['Churned', 'Not Churned', 'Churned', 'Churned', 'Not Churned', 'Churned']
})


percentage = data['churn_status'].value_counts(normalize=True) * 100


plt.figure(figsize=(6, 6))  
percentage.plot(kind='pie', autopct='%1.2f%%', startangle=90)
plt.title('Churn Status Distribution (Percentage)')
plt.ylabel('') 
plt.show()
```

## Line Chart

```python
data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [100, 120, 130, 90, 110, 150]
})
```

```python
plt.figure(figsize=(6, 4))
data.set_index('Month')['Sales'].plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
sns.lineplot(x='Month', y='Sales', data=data, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

## Histogram

```python
data = pd.DataFrame({
    'Income': [2500, 2700, 3000, 3200, 2800, 2600, 4000, 4200, 3800, 2900]
})
```

```python
num_col = 'Income'
```

```python
plt.figure(figsize=(6, 4))
data['Income'].plot(kind='hist', bins=5)
plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(data[num_col], bins=5, kde=False, color='skyblue')
```
