```python
data = pd.read_csv("C:/Users/Chet/Documents/courses_data.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cross_tab = pd.crosstab(data['Course Name'], data['Credits'])

plt.figure(figsize=(8, 5)) 
cross_tab.plot(kind='bar', stacked=False)
plt.title('Course Name by Credits')
plt.xlabel('Course Name')
plt.ylabel('Count')
plt.show()
```

## Stacked or Clustered

```python
cross_tab = pd.crosstab(data['Course Level'], data['Course Name'])
```

```python
plt.figure(figsize=(10, 6))
cross_tab.plot(kind='bar', stacked=False)
plt.title('Course Distribution by Level and Name')
plt.xlabel('Course Level')
plt.ylabel('Number of Courses')
plt.legend(title='Course Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='Course Level', hue='Course Name', data=data)
```

## Pie Chart

```python
credit_counts = data['Credits'].value_counts()
```

```python
plt.figure(figsize=(6, 6))
credit_counts.plot(kind='pie', autopct='%1.2f%%')
plt.title('Percentage Distribution of Courses by Credit Value')
plt.ylabel('') 
plt.show()
```

## Line chart

```python
group_by = data.groupby('Course Level')['Credits'].mean()
```

```python
plt.figure(figsize=(6, 4))
group_by.plot(kind='line', marker='o')
plt.title('Average Credits by Course Level')
plt.xlabel('Course Level')
plt.ylabel('Average Credits')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.lineplot(x='Course Level', y='Credits', data=data, marker='o')
```

## Histogram

```python
plt.figure(figsize=(6, 4))
data['Credits'].plot(kind='hist', bins=5, edgecolor='black')
plt.title('Distribution of Credits')
plt.xlabel('Credits')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(data['Credits'], bins=5, kde=False, color='skyblue', edgecolor='black')
```
