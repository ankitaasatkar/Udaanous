####  Loc and Iloc

```python
data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob'],
    'Salary': [50000, 60000, 50000, 70000, 60000]})
```

####  Combining and condition

```python
data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Age': [25, 35, 45, 28],
    'Salary': [50000, 60000, 70000, 48000]})
```

```python
filtered_data = data[(data['Age'] > 30) | (data['Salary'] < 50000)]
filtered_data
```

```python
filtered_data = data[(data['Age'] > 20) & (data['Salary'] < 60000)]
filtered_data
```

## 14 Joining 

```python
data = pd.DataFrame({
    'ID': [1,2,3,4],
    'name': ['Alice', 'Bob', 'Charlie', 'Bob'],
    'salary': [50000, 60000, 70000, 60000]
})
```

```python
data2 = pd.DataFrame({
    'order_id': [2, 3, 4, 5],
    'ID': [1, 2, 3, 4],
    'order_date': ['2024-05-01', '2024-05-03', '2024-05-05', '2024-05-07']
    
})
```

```python
joined_data = pd.merge(data,data2, on='ID', how='inner')
joined_data.head()
```

## 15 concatenating

```python
data = pd.DataFrame({
    'ID': [1,2,3,4],
    'name': ['Alice', 'Bob', 'Charlie', 'Bob'],
    'salary': [50000, 60000, 70000, 60000]
})
```
